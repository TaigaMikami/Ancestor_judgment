```python
from keras.models import load_model
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib import image as img_mat
```

```python
# CIFAR-10公式ラベル
cifar10_labels = np.array([
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'])
```

```python
model = load_model(os.path.join('results','my_model.h5')) # 保存したモデルから読み込む
```

```python
img_path = 'images/gutty.png' # 任意の画像
img_gutty = img_mat.imread(img_path)
plt.imshow(img_gutty)
plt.show()
```

```python
img_path = 'images/gutty.png' # 任意の画像
img = image.load_img(img_path, target_size=(32, 32)) #入力画像のサイズ
print(img)
plt.imshow(img)
plt.show()
x = image.img_to_array(img) # 画像データをnumpy.arrayへ変換
print(x[0])
print(x[0].astype(np.uint8))
plt.imshow(x.astype(np.uint8))
plt.show()
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
preds = model.predict(x) # modelは学習させたもの
category = np.argmax(preds, axis=1)
```

```python
classed = category[0]
print(preds[0][classed])
print(category)
```

```python
print(x[0].shape)
plt.imshow(x[0])
plt.show()
```

```python
print('分類されたものは{0}です。' .format(cifar10_labels[classed]))
print('精度:{0}です。' .format(preds[0][classed]*100))
```
